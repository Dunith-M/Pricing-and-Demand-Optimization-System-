# Handling Missing Values

```python
missing = df.isnull().sum().sort_values(ascending=False)
missing_pct = (missing / len(df)) * 100

missing_df = pd.DataFrame({
    "missing_count": missing,
    "missing_percent": missing_pct
})

missing_df.head(20)
```

### Dropped Columns and Justification

1. **license**
   - Contains ~100% missing values.
   - Provides no usable signal for modeling.
   - Including this feature would introduce noise without predictive value.

2. **house_rules**
   - High missing rate (~50%) and unstructured text format.
   - Requires advanced NLP processing, which is out of scope for the current phase.
   - Low direct impact on pricing and demand modeling relative to complexity.

3. **host_name**
   - Identifier-type feature with no intrinsic predictive meaning.
   - Does not generalize across listings and may introduce unnecessary variance.

4. **NAME**
   - Listing title (textual feature) requiring NLP for meaningful extraction.
   - High dimensionality and noise risk without proper text processing.
   - Not critical for baseline pricing and demand models.

5. **country_code**
   - Redundant feature since `country` already captures the same information.
   - Keeping both would introduce multicollinearity without added value.

6. **construction_year**
   - Potentially useful but often noisy/inconsistent in real-world datasets.
   - Requires transformation (e.g., property age) and validation.
   - Dropped at this stage to maintain data reliability and simplicity.

### Missing Value Imputation Strategy

The following numerical features contain a very small percentage of missing values (<1%). 
Given their importance in pricing and demand modeling, we apply **median imputation**.

#### Rationale for Median Imputation

1. **availability_365**
   - Represents the number of days a listing is available per year.
   - Can contain skewed values (e.g., always available vs rarely available).
   - Median is robust to outliers and preserves distribution.

2. **minimum_nights**
   - Indicates booking constraints set by hosts.
   - Typically right-skewed (few listings have very high minimum nights).
   - Median avoids distortion from extreme values.

3. **review_rate_number**
   - Reflects user ratings (trust/quality signal).
   - Important for demand prediction.
   - Median maintains central tendency without bias from outliers.

4. **calculated_host_listings_count**
   - Proxy for host experience or professionalism.
   - Distribution is usually skewed (few hosts own many listings).
   - Median prevents over-influence from large commercial hosts.

5. **price**
   - Core variable for pricing and revenue modeling.
   - Must not be dropped due to missing values.
   - Median imputation ensures stability without introducing bias from extreme listings.

#### Summary

- Missing percentage is very low → safe to impute.
- Median is chosen due to:
  - robustness to skewness
  - resistance to outliers
  - preservation of realistic distribution

```python
cols_median = [
    "availability 365",
    "minimum nights",
    "review rate number",
    "calculated host listings count"
]

for col in cols_median:
    df[col].fillna(df[col].median())

print("Missing values after imputation:")
print(df[cols_median].isnull().sum())
```

```python
# Clean and convert price
df["price"] = (
    df["price"]
    .replace(r"[\$,]", "", regex=True)
    .astype(float)
)

# Clean and convert service_fee
df["service fee"] = (
    df["service fee"]
    .replace(r"[\$,]", "", regex=True)
    .astype(float)
)

# Fill missing values with median
df["price"].fillna(df["price"].median(), inplace=True)
df["service fee"].fillna(df["service fee"].median(), inplace=True)

# Check
print(df[["price", "service fee"]].isnull().sum())
print(df[["price", "service fee"]].head())
```

```python
# Check missing values for specific columns
print("Missing values count:")
print(df[["price", "service fee"]].isnull().sum())

# Also check percentage (more informative)
missing_pct = df[["price", "service fee"]].isnull().mean() * 100
print("\nMissing values percentage:")
print(missing_pct)
```

```python
cols_mode = [
    "instant_bookable",
    "neighbourhood group"
]

for col in cols_mode:
    mode_value = df[col].mode()[0]
    df[col].fillna(mode_value, inplace=True)

# Check
print(df[cols_mode].isnull().sum())
```

```python
# Standardize values first
df["cancellation_policy"] = df["cancellation_policy"].str.lower().str.strip()

# Get mode
mode_value = df["cancellation_policy"].mode(dropna=True)[0]

# Fill missing values
df["cancellation_policy"] = df["cancellation_policy"].fillna(mode_value)

# Check
print(df["cancellation_policy"].isnull().sum())
print("Mode used:", mode_value)
```
