```python
import pandas as pd
import json
import os
```

```python
DATA_PATH = "../data/raw/airbnb.csv"

df = pd.read_csv(DATA_PATH)
```

```python
df = df.dropna()

df["demand_score"] = df["reviews per month"] / (df["availability 365"] + 1)
```

#Important Columns

```python
strong_features = [
    "reviews per month",
    "availability 365",
    "room type",
    "neighbourhood group",
    "host_identity_verified",
    "review rate number",
    "calculated host listings count"
]

weak_features = [
    "id",
    "host id",
    "last review"
]

potential_features = [
    "price",
    "service fee",
    "minimum nights",
    "instant_bookable",
    "cancellation_policy",
    "neighbourhood",
    "lat",
    "long"
]
```

#Lets Build the Hypothesis

```python
hypotheses = [
    "Higher price reduces demand_score",
    "Lower availability indicates higher demand",
    "Entire homes have higher revenue potential",
    "Better host credibility increases demand",
    "Location significantly affects pricing and demand",
    "Minimum nights constraint affects booking probability"
]
```

```python
summary = {
    "strong_features": strong_features,
    "weak_features": weak_features,
    "potential_features": potential_features,
    "hypotheses": hypotheses
}
```

```python
OUTPUT_JSON = "../artifacts/reports/eda/feature_summary.json"

os.makedirs(os.path.dirname(OUTPUT_JSON), exist_ok=True)

with open(OUTPUT_JSON, "w") as f:
    json.dump(summary, f, indent=4)

print("Saved feature summary")
```

```python
report = f"""
# EDA REPORT — Airbnb Pricing Optimization

## 1. Strong Features

{', '.join(strong_features)}

## 2. Weak Features

{', '.join(weak_features)}

## 3. Potential Features

{', '.join(potential_features)}

## 4. Hypotheses

""" + "\n".join([f"- {h}" for h in hypotheses])

OUTPUT_MD = "../artifacts/reports/eda/eda_report.md"

with open(OUTPUT_MD, "w") as f:
    f.write(report)

print("Saved EDA report")
```
